# Data  and Preprocessing

```python
#Plot
import matplotlib.pyplot as plt
import seaborn as sns

#Math packages
import pandas as pd
import numpy as np
import math

#Graph
import networkx as nx

#linear programming
import cvxpy as cp
import gurobipy as gp
```

```python
languageDF = pd.read_excel('Frances_Material/MaximizingTeamDiversityData.xlsx', sheet_name='Language').drop('Languages', axis=1)
leaderDF = pd.read_excel('Frances_Material/MaximizingTeamDiversityData.xlsx', sheet_name='Leadership').drop('Lead. Styles', axis=1)
learningDF = pd.read_excel('Frances_Material/MaximizingTeamDiversityData.xlsx', sheet_name='Learning').drop('Learn. Styles', axis=1)
likesDF = pd.read_excel('Frances_Material/MaximizingTeamDiversityData.xlsx', sheet_name='Like')
dislikesDF = pd.read_excel('Frances_Material/MaximizingTeamDiversityData.xlsx', sheet_name='Dislikes')

studentDF = languageDF.join(leaderDF).join(learningDF).join(likesDF).join(dislikesDF).drop(['Students'])
```

```python
#undoing unessecary 1-hot encoding

studentDF['leadershipStyle'] = studentDF.filter(['Amiable','Expressive','Analytical','Driver']).idxmax(1)
studentDF['learningStyle'] = studentDF.filter(['ref-int-vrb-seq','act-int-vrb-seq', 'ref-sns-vrb-seq',
 'act-sns-vrb-seq',
 'ref-int-vis-seq',
 'act-int-vis-seq',
 'ref-sns-vis-seq',
 'act-sns-vis-seq',
 'ref-int-vrb-glo',
 'act-int-vrb-glo',
 'ref-sns-vrb-glo',
 'act-sns-vrb-glo',
 'ref-int-vis-glo',
 'act-int-vis-glo',
 'ref-sns-vis-glo',
 'act-sns-vis-glo']).idxmax(1)
studentDF = studentDF.drop(['Amiable','Expressive','Analytical','Driver','ref-int-vrb-seq','act-int-vrb-seq', 'ref-sns-vrb-seq',
 'act-sns-vrb-seq',
 'ref-int-vis-seq',
 'act-int-vis-seq',
 'ref-sns-vis-seq',
 'act-sns-vis-seq',
 'ref-int-vrb-glo',
 'act-int-vrb-glo',
 'ref-sns-vrb-glo',
 'act-sns-vrb-glo',
 'ref-int-vis-glo',
 'act-int-vis-glo',
 'ref-sns-vis-glo',
 'act-sns-vis-glo'], axis=1)
```

```python
likeColumn = []
dislikeColumn = []

for i in studentDF.loc[:, 'Student-1Like':'Student-96Like']:
    likeColumn.append(studentDF[i].sum())
    
for i in studentDF.loc[:, 'Student-1Dislikes':'Student-96Dislikes']:
    dislikeColumn.append(studentDF[i].sum())

studentDF['totalLikes'] = likeColumn
studentDF['totalDislikes'] = dislikeColumn
studentDF['likeDelta'] = studentDF['totalLikes'] - studentDF['totalDislikes']
```

# Graph Building

```python
# Graph time! (this is student likes)
L = nx.DiGraph()
L.add_nodes_from(studentDF.index.values)

for i in studentDF.loc[:, 'Student-1Like':'Student-96Like']:
    for j in studentDF.index:
        if(studentDF.loc[j][i] == 1):
            L.add_edge(j, i[:-4])
```

```python
# Graph time! (this is a directed student dislikes)
D = nx.DiGraph()
D.add_nodes_from(studentDF.index.values)

for i in studentDF.loc[:, 'Student-1Dislikes':'Student-96Dislikes']:
    for j in studentDF.index:
        if(studentDF.loc[j][i] == 1):
            D.add_edge(j, i[:-4])
```

```python
# Graph time! (this is undirected student dislikes)
D_U = nx.Graph()
D_U.add_nodes_from(studentDF.index.values)

for i in studentDF.loc[:, 'Student-1Dislikes':'Student-96Dislikes']:
    for j in studentDF.index:
        if(studentDF.loc[j][i] == 1):
            D_U.add_edge(j, i[:-8])
```

```python
#problem_formulation with GUROBI solver

#Decision Variable
X = cp.Variable((num_students, num_clusters), boolean=True) #, integer=True)
row_ones = np.ones((num_clusters, 1))
num_ones = np.ones((num_students, 1))

#Constraints
constraints = [] #cp.sum(Y, axis=1) == 1]
#constraints.append(cp.sum(X, axis=1) >= 1)
constraints.append(cp.matmul(X,row_ones) == num_ones)
constraints.append(cp.sum(X, axis=0) >= min_cluster)
constraints.append(cp.sum(cp.diag(cp.matmul(X.T,cp.matmul(dislike_array,X)))) <= 100)

#actual objective
#obj = cp.Minimize(cp.sum(cp.diag(cp.matmul(X.T,cp.matmul(dislike_array,X)))))

#linear objective
obj = cp.Minimize(cp.sum(cp.matmul(dislike_array,X)))
cluster_prob_4 = cp.Problem(obj, constraints)
cluster_prob_4.solve(solver='GUROBI')
```

```python
#problem_formulation with GUROBI solver, with non-matrix objective to avoid DCP issue?

#Decision Variable
X = cp.Variable((num_students, num_clusters), boolean=True) #, integer=True)
row_ones = np.ones((num_clusters, 1))
num_ones = np.ones((num_students, 1))

#Constraints
constraints = [] #cp.sum(Y, axis=1) == 1]
#constraints.append(cp.sum(X, axis=1) >= 1)
constraints.append(cp.matmul(X,row_ones) == num_ones)
constraints.append(cp.sum(X, axis=0) >= min_cluster)

#actual objective
#obj = cp.Minimize(cp.sum(cp.diag(cp.matmul(X.T,cp.matmul(dislike_array,X)))))

#linear objective
objective = []
for i in range (0, num_students):
    for j in range (i+1, num_students):
        objective.append(cp.sum(X[i][0] * X[j][0] * dislike_array[i][j]))
#objective = [cp.sum(dislike_array[i,j]*X[i,k]*X[j,k]) for (i,j,k) in zip(range(num_students), range(num_students), range(num_clusters))]
obj = cp.Minimize(cp.sum(objective))

cluster_prob_4 = cp.Problem(obj, constraints)
cluster_prob_4.solve(solver='GUROBI')
```

## Just using Gurobipy

```python
dislike_DF = symmetricDislike(studentDF.loc[:, 'Student-1Dislikes':'Student-96Dislikes'])
num_students = len(dislike_DF.index.values)
dislike_array = dislike_DF.values

#manual inputs
num_clusters = 4 #number of clusters
min_cluster = 20 #min cluster size
```

```python
m = gp.Model('Dislike')

row_ones = np.ones((num_clusters, 1))
num_ones = np.ones((num_students, 1))

teams = m.addVars(num_students, num_clusters, vtype=gp.GRB.BINARY, name='G')
m.addConstrs((teams.sum(i,'*') == 1) for i in range(num_students))
m.addConstrs((teams.sum('*',i) >= min_cluster) for i in range(num_clusters))

#model.addConstrs((vars.sum(i,'*',v) == 1
#                 for i in range(n)
#                 for v in range(n)), name='R')

objective = gp.QuadExpr()
for i in range (0, num_students):
    for j in range (i+1, num_students):
        objective.add(teams[i,1] * teams[j,1], dislike_array[i,j])
        
m.setObjective(objective, gp.GRB.MINIMIZE)

m.optimize()

solution = m.getAttr('x', teams)
```

```python
def gp2array(input):
    #takes a gurobipy solution as a string, returns an nparray
    process = input[:-1].split(", (")
    for i in range(0, len(process)):
        process[i] = int(process[i][-3:-2])
        
    final_array = np.empty([num_students, num_clusters])
    for i in range(0, num_students):
        final_array[i,0] = process[i*4]
        final_array[i,1] = process[(i*4) + 1]
        final_array[i,2] = process[(i*4) + 2]
        final_array[i,3] = process[(i*4) + 3]
    return( process)
    
```
